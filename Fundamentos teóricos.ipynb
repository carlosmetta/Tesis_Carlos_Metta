# Descomposición de valores singulares ejemplo


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz original A
A = np.array([
    [25, 70, 175, 0], 
    [75, 65, 160, 1], 
    [55, 110, 185, 0], 
    [20, 55, 170, 1], 
    [15, 50, 172, 0]
])

# Calcular la descomposición en valores singulares
U, S, VT = np.linalg.svd(A)

# Paso 3: Proyectar las personas sobre los primeros dos componentes
# Multiplicamos U[:, :2] por diag(S[:2]) para obtener la representación en 2D
U2 = U[:, :2]
S2 = np.diag(S[:2])
projected_2D = U2 @ S2

# Paso 4: Graficar
plt.figure(figsize=(7, 6))
plt.scatter(projected_2D[:, 0], projected_2D[:, 1], c='skyblue', edgecolors='black', s=100)

# Añadir etiquetas de cada persona
for i in range(projected_2D.shape[0]):
    plt.text(projected_2D[i, 0] + 1, projected_2D[i, 1], f'Persona {i+1}', fontsize=11)

# Etiquetas y formato
plt.title('Proyección de las personas en los dos primeros componentes (DVS)')
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.grid(True)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
